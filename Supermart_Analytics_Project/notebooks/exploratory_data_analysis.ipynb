```python
# Cell 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 2: Load Cleaned Data
# Since this notebook is in the notebooks folder, if the cleaned data is in the data folder at the project root,
# use the correct relative path:
data = pd.read_csv('../data/cleaned_supermart_data.csv')
print("âœ… Data Overview:")
print(data.head())

# Cell 3: Sales Distribution Plot
plt.figure(figsize=(8, 5))
sns.histplot(data['Sales'], bins=30, kde=True, color='skyblue')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

# Cell 4: Sales by Category (Bar Chart)
plt.figure(figsize=(10, 6))
# Get the count of each category (assuming category is encoded as numbers)
category_counts = data['Category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Sales by Category (Count)')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Cell 5: Yearly Sales Trend (Line Plot)
plt.figure(figsize=(10, 6))
# Group the data by Order_Year and sum up the Sales
yearly_sales = data.groupby('Order_Year')['Sales'].sum().reset_index()
sns.lineplot(x='Order_Year', y='Sales', data=yearly_sales, marker='o', color='coral')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

print("âœ… Exploratory Data Analysis Completed!")
```
